### Getting the number of outliers: define the upper bound and lower bound for the 'fare' column at 95% or 5% and count how many passengers are considered outliers

```python
#Answer here

# Calculate the 95th and 5th percentiles for the 'fare' column
percentile_95 = df['fare'].quantile(0.95)
percentile_5 = df['fare'].quantile(0.05)

# Define the upper and lower bounds for outliers
upper_bound = percentile_95
lower_bound = percentile_5

# Count the number of passengers with fares above the upper bound or below the lower bound
outliers_count = ((df['fare'] > upper_bound) | (df['fare'] < lower_bound)).sum()

# Print the number of outliers
print("Number of outliers in the 'fare' column:", outliers_count)
```

### Replace outliers in the dataframe by the mean of the 'fare' column

```python
#Answer here

# Calculate the mean of the 'fare' column
fare_mean = df['fare'].mean()

# Calculate the 95th and 5th percentiles for the 'fare' column
percentile_95 = df['fare'].quantile(0.95)
percentile_5 = df['fare'].quantile(0.05)

# Replace outliers with the mean of the 'fare' column
df['fare'] = df['fare'].apply(lambda x: fare_mean if x > percentile_95 or x < percentile_5 else x)

# Print the modified DataFrame
print(df)
```

# Filling missing data in 'age' column by using an educated guess:

### Write a function that takes two columns, 'age' and 'pclass', as input and depending on which pclass, if the age is NaN (command: pd.isnull(age)   ) it returns age 40, 30 or 25. For example, if input is [NaN, 1], it returns age = 40

```python

def fill_age(age, pclass):
    """
    This function takes two columns, 'age' and 'pclass', as input and depending on which pclass, 
    if the age is NaN it returns age 40, 30 or 25. For example, if input is [NaN, 1], it returns age = 40
    
    Parameters:
    age (float): Age of the passenger
    pclass (int): Passenger class (1, 2, or 3)
    
    Returns:
    float: Age of the passenger after filling missing values
    """
    if pd.isnull(age):
        if pclass == 1:
            return 40
        elif pclass == 2:
            return 30
        else:
            return 25
    else:
        return age
```

### Use apply to clean up age database for NaN using the function created above

```python
#Answer here
df_combined_drop['age'] = df_combined_drop.apply(lambda row: fill_age(row['age'], row['pclass']), axis=1)
print(df_combined_drop)
# you can see the age has changed from Nan to 25 in row 1314.
```

```python
#Answer here
# Define a function to convert 'sex' to 0 or 1
def convert_sex_to_binary(sex):
    """
    This function takes a string 'sex' as input and returns 0 if sex is 'male' and 1 if sex is 'female'.
    
    Parameters:
    sex (str): Sex of the passenger
    
    Returns:
    int: 0 if sex is 'male', 1 if sex is 'female'
    """
    if sex == 'male':
        return 0
    else:
        return 1

# Apply the function to the 'sex' column
df_combined_drop['sex'] = df_combined_drop['sex'].apply(convert_sex_to_binary)

# Print the modified DataFrame
print(df_combined_drop)
```
