```python
import pandas as pd
import datetime
from random import randint
from time import sleep
import os
```

```python
url = 'https://www.hockey-reference.com/leagues/NHL_2024_games.html'

dfs = pd.read_html(url)
df = dfs[0]
```

```python
dates = pd.to_datetime(df["Date"],format="%Y-%m-%d").dt.date
dates = pd.Series(dates).drop_duplicates().tolist()
# dates
```

```python
nhl_url = ''
end_date = dates[-1]
output_path = r"C:\Users\roser\OneDrive\HB\St Marys\Classes\BIA 650 Data Mining for Decision Making\Week 3\OutputScrape.csv"
```

```python
# loop through each date in the dates list
for date in (dates):
# populate the nhl_url with our dynamic url. 
# Notice that we are now using the `date` 
# to populate fd and td params.
# fd={date}&td={date}
	nhl_url = f'https://www.naturalstattrick.com/teamtable.php?fromseason=20212022&thruseason=20212022&stype=2&sit=ev&score=all&rate=y&team=all&loc=B&gpf=410&fd={date}&td={date}'
	# if the date is earlier or equal to the `end_date`, proceed
	if(date <= end_date):
		nhl_dfs = pd.read_html(nhl_url, index_col=0)
		nhl_df = nhl_dfs[0]
		# add a Date column based on the game date
		nhl_df['Date'] = date
		nhl_df.to_csv(output_path, mode='a',header = not os.path.exists(output_path), index=False)
		sleep(randint(10,25))
print("Complete")
```
